```python
final=pd.read_csv('/Main.csv')
final.set_index('Ticker', inplace=True)
final= final[final['Industry']!= 'Asset Management']
final= final[final['Sector']!= 'Finance and Insurance']
final= final[final['Sector']!= 'Financial Services']
final= final[final['Sector']!= 'Real Estate']
final= final[final['Sector']!= 'Real Estate and Rental and Leasing']
final["Business"] = final["Industry"] + final["Sector"]
final.head()
```

```python
tickers = final.index.tolist()
print(tickers)
```

```python
from yahooquery import Ticker
faang = Ticker(tickers)

df= pd.DataFrame(faang.summary_detail)
```

```python
df1= df.T
df.index.name = 'Ticker'
df1= df1[['open','volume','marketCap','fiftyTwoWeekLow','averageVolume10days','previousClose','beta']]
df1 = pd.concat([df1, final], axis=1)
df1 = df1.loc[df1["marketCap"].apply(lambda x: str(x).isdigit())]
df1["marketCap"] = df1["marketCap"].astype(int)
df1 = df1[df1["marketCap"] > 300000000]
df1['Price_Change']=(df1.open- df1.previousClose) / df1.previousClose* 100
df1['low_52_Week']= ((df1.open- df1.fiftyTwoWeekLow) / df1.fiftyTwoWeekLow)*100
df1= df1[df1['averageVolume10days']!= 0]
df1['Volume_Change']= (df1.volume- df1.averageVolume10days)/df1.averageVolume10days*100
df1 = df1.sort_values(by=['low_52_Week'], ascending=True)
df_Lowprice= df1[df1['low_52_Week'] <5]
df_Lowprice1= df1[df1['low_52_Week'] <20]
df_Volume_Change= df1.sort_values(by=['Volume_Change'], ascending=False)
df_Price_Change = df1.sort_values(by=['Price_Change'], ascending=True)
df_Price_Change = df_Price_Change[df_Price_Change['Price_Change'] <-9]
df_Price_Change1= df_Price_Change.head(30)
df1['Target']= df1.volume- df1.averageVolume10days
df_Target= df1[df1['Target'] >0]
df_Target= df_Target[df_Target['beta'].apply(lambda x: x['value'] if isinstance(x, dict) else x) >=1]
df_Target= df_Target.head(25)
df_all=pd.concat([df_Price_Change1,df_Volume_Change,df_Lowprice,df_Target])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# assume we have a dataframe 'df' with columns 'Industry', 'Current Volume', and 'Avg Volume Last 10 Days'

# group by industry and calculate the average volume for each industry
industry_avg_volume = df1.groupby('Business')['averageVolume10days'].mean()

# calculate the difference between current volume and average volume for each stock
df1['Volume Difference'] = df1['volume'] - df1.groupby('Business')['averageVolume10days'].transform('mean')

# plot the results
plt.figure(figsize=(200, 6))
plt.bar(industry_avg_volume.index, df1.groupby('Business')['Volume Difference'].mean())
plt.axhline(y=0, color='black', linestyle='--')
plt.xticks(rotation=45)
plt.xlabel('Business')
plt.ylabel('Volume Difference')
plt.title('Difference in Current Volume vs Avg Volume Last 10 Days by Industry')
plt.show()
```

```python
import requests

def get_realtime_stock_price(ticker):
    url = f"https://cloud.iexapis.com/stable/stock/{ticker}/quote?token=pk_8f56852f68c146349941fa9f3bf51de6"
    response = requests.get(url)
    data = response.json()
    return float(data["latestPrice"])

results = []
for ticker in tickers:
    realtime_price = get_realtime_stock_price(ticker)
    results.append({"Ticker": ticker, "Price": realtime_price})

Market_Price = pd.DataFrame(results)
```
